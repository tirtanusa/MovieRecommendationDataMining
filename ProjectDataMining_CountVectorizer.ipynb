# Pre-Processing

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
stop_words = set(stopwords.words('english')) 

lemmatizer = WordNetLemmatizer()

def process_text(text):
    # Tokenisasi teks
    tokens = word_tokenize(text)
    # Penghapusan stopwords dan lemmatisasi
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return tokens

df['Descrption_Processed'] = df['Descrption'].apply(lambda x:process_text(str(x)))

df
```

# Vectorizing

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def join_words(words):
    if isinstance(words,list):
        return ' '.join(words)
    return words

df['Descrption_vect'] = df['Descrption_Processed'].apply(join_words)

vectorizer = CountVectorizer()
vec_desc = vectorizer.fit_transform(df['Descrption_vect'])



similarity_matrix = cosine_similarity(vec_desc)

similarity_matrix
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


vectorizer = TfidfVectorizer(stop_words = 'english',lowercase = True)
x = vectorizer.fit_transform(df['Descrption_vect'])


tfidf = pd.DataFrame(x.toarray(), columns = vectorizer.get_feature_names_out())

similarity_matrix_tfidf = cosine_similarity(tfidf)
```

# Movie Recommendation (OUTPUT)

```python
def recommend_movies(movie_title, df, similarity_matrix, top_n=5):
    # Cari indeks film dalam DataFrame
    if movie_title in df['Title'].values:
        movie_idx = df.index[df['Title'] == movie_title].tolist()[0]
    else:
        return "Film tidak ditemukan."

    # Dapatkan skor kesamaan untuk film tersebut terhadap semua film lain
    similarity_scores = list(enumerate(similarity_matrix[movie_idx]))

    # Urutkan film berdasarkan skor kesamaan
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Dapatkan indeks dari 5 film paling mirip (tidak termasuk film itu sendiri)
    top_movie_indices = [idx for idx, score in similarity_scores[1:top_n+1]]

    # Kembalikan judul film
    top_movies = df.iloc[top_movie_indices]
    return top_movies

# Contoh penggunaan fungsi
input_movie = "Avengers: Endgame"
recommended_movies = recommend_movies(input_movie, df, similarity_matrix)
recommended_movies
```
